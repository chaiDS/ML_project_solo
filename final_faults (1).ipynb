# Data Transform

```python
#Data transform using robust tranform
from sklearn.preprocessing import RobustScaler
from pandas import DataFrame
numerical_data = faults_.iloc[:,:-1]
#performing robust scaler transform on dataset
trans = RobustScaler()
data = trans.fit_transform(numerical_data)
dataset = DataFrame(data)
print(dataset.shape)
print(dataset)
```

```python
X = dataset.iloc[:,:] 
print(X)
```

```python
from sklearn.preprocessing import LabelEncoder
y = faults_.iloc[:,-1]
y = LabelEncoder().fit_transform(y.astype('str'))
y = DataFrame(y)
print(y)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y, train_size = 0.8, random_state=100)
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
#Test options and evaluation metric
num_folds = 10
seed = 7
scoring = 'accuracy'
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
```

```python
# Spot-Check Algorithms
models = []
models.append(('LR', LogisticRegression()))
models.append(('LDA', LinearDiscriminantAnalysis()))
models.append(('KNN', KNeighborsClassifier()))
models.append(('CART', DecisionTreeClassifier()))
models.append(('NB', GaussianNB()))
models.append(('SVM', SVC()))
```

```python
import warnings
warnings.filterwarnings("ignore")
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
results = []
names = []
for name, model in models:
    kfold = KFold(n_splits=num_folds, random_state=seed)
    cv_results = cross_val_score(model, X_train, y_train, cv=kfold, scoring=scoring)
    results.append(cv_results)
    names.append(name)
    msg = "%s: %f (%f)" % (name, cv_results.mean(), cv_results.std())
    print(msg)
```

---The results suggests that ***KNN*** may be worth further

These are just mean accuracy values. It is always wise to look at the distribution of accuracy values calculated across cross validation folds. We can do that graphically using box and whisker plots.

```python
#Compare Algorithms
fig = plt.figure()
fig.suptitle('Algorithm Comparison')
ax = fig.add_subplot(111)
plt.boxplot(results)
ax.set_xticklabels(names)
plt.show()
```

```python
k_range = list(range(1,50))
weight_options = ['uniform', 'distance']
metric_K = ["euclidean","manhattan"]
param_grid = dict(n_neighbors = k_range, weights = weight_options, metric = metric_K)
```

```python
knn = KNeighborsClassifier()
```

```python
from sklearn.model_selection import GridSearchCV
grid = GridSearchCV(knn, param_grid, cv = 8,verbose = 1, n_jobs = -1)
grid.fit(X,y)
```

```python
print (grid.best_score_)
print (grid.best_params_)
print (grid.best_estimator_)
```

```python
#Training and Predicting
from sklearn.neighbors import KNeighborsClassifier
knnmodel = KNeighborsClassifier(n_neighbors = 33)
knnmodel.fit(X_train,y_train)
ypredict = knnmodel.predict(X_test)
```

```python
#Accuracy
from sklearn.metrics import accuracy_score
acc = accuracy_score(y_test,ypredict)
acc
```

```python
#Confusion Matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,ypredict)
cm
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,ypredict))
```
